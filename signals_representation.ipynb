```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve, freqz, impulse, step
from scipy.fftpack import fft, fftshift
```

```python
#Continuous and Discrete Signal Representation
def continuous_signal(t):
    return np.sin(2 * np.pi * 1 * t)  # Sine wave with frequency 1Hz

def discrete_signal(n):
    return np.sin(2 * np.pi * 0.1 * n)  # Discrete sine wave

t = np.linspace(0, 2, 1000)
n = np.arange(0, 20)
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(t, continuous_signal(t))
plt.title("Continuous Signal")
plt.subplot(1, 2, 2)
plt.stem(n, discrete_signal(n))
plt.title("Discrete Signal")
plt.show()
```

```python
#Basic Operations (Addition, Scaling, Multiplication)
s1 = continuous_signal(t)
s2 = 0.5 * continuous_signal(t)
plt.figure(figsize=(6, 4))
plt.plot(t, s1, label='Original')
plt.plot(t, s2, label='Scaled by 0.5', linestyle='dashed')
plt.legend()
plt.title("Signal Scaling")
plt.show()
```

```python
#Time Transformations (Shifting, Scaling, Reversal)
shifted_signal = continuous_signal(t - 0.5)  # Shifted by 0.5 sec
reversed_signal = continuous_signal(-t)  # Time reversal
plt.figure(figsize=(6, 4))
plt.plot(t, s1, label='Original')
plt.plot(t, shifted_signal, label='Shifted by 0.5 sec', linestyle='dashed')
plt.legend()
plt.title("Time Shift")
plt.show()
```

```python
#Fourier Transform Analysis
X = fft(continuous_signal(t))
frequencies = np.fft.fftfreq(len(t), d=t[1]-t[0])
plt.figure(figsize=(6, 4))
plt.plot(fftshift(frequencies), fftshift(abs(X)))
plt.title("Magnitude Spectrum (Fourier Transform)")
plt.show()
```

```python
#Impulse and Step Response
system = ([1], [1, 1])  # Simple system H(s) = 1 / (s + 1)
t_imp, y_imp = impulse(system)
t_step, y_step = step(system)
plt.figure(figsize=(6, 4))
plt.plot(t_imp, y_imp, label="Impulse Response")
plt.plot(t_step, y_step, label="Step Response")
plt.legend()
plt.title("System Responses")
plt.show()
```

```python
#Convolution (Discrete Signal Processing)
x = np.array([1, 2, 3, 4])  # Sample discrete signal
h = np.array([0.2, 0.5, 0.2])  # Sample filter
conv_result = convolve(x, h, mode='full')
plt.figure(figsize=(6, 4))
plt.stem(conv_result)
plt.title("Convolution of x[n] and h[n]")
plt.show()
```
